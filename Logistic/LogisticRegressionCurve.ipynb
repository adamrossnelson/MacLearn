```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

xvar = np.linspace(0, 6000, 1000)
yvar = 1 / (1 + np.exp( - (5 + xvar * -.0025)))

df = pd.DataFrame({'xvar':xvar, 'yvar':yvar})
```

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)
ax.plot('xvar', 'yvar', data=df)
ax.set(title='Purchasing Predictions',
       ylabel='Probability', 
       xlabel='Time In Number of Days')
```
